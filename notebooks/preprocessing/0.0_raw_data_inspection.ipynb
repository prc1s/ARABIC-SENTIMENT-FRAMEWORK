```python
import xml.etree.ElementTree as ET
from pathlib import Path
from collections import Counter

def inspect(path:Path):
    tree = ET.parse(path)
    root = tree.getroot()

    reviews = root.findall("Review")
    sentences = root.findall(".//sentence")
    opinions = root.findall(".//Opinion")

    categories = Counter()
    polarities = Counter()

    for op in opinions:
        categories[op.attrib.get("category")] += 1
        polarities[op.attrib.get("polarity")] += 1

    return {
        "reviews": len(reviews),
        "sentences": len(sentences),
        "opinions": len(opinions),
        "categories": categories,
        "polarities": polarities
    }

sb2_test_stats = inspect("raw_datasets/SemEval2016_arabic/SemEval2016_arabic/AR_HOTE_SB2_TEST.xml.gold")
sb2_stats = inspect("raw_datasets/SemEval2016_arabic/SemEval2016_arabic/AR_Hotels_Train_SB2.xml")
print("SB2 Test Inspection")
for k, v in sb2_test_stats.items():
    print(k, ":", v)
print("SB2 Train Inspection")
for k, v in sb2_stats.items():
    print(k, ":", v)

```
